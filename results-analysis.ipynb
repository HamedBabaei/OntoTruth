```python
import pandas as pd

llms = ['Qwen2.5-0.5B-Instruct', 'Llama-3.3-70B-Instruct', 'Llama-3.1-8B-Instruct', 'Qwen2.5-72B-Instruct', 'Qwen2.5-7B-Instruct', 'Llama-3.2-1B-Instruct']

flaw_q = [14, 15, 17, 44, 47, 48, 49, 50, 58]

dfs = {llm: pd.read_csv(f"results/{llm}-evaluated.csv")[~pd.read_csv(f"{llm}-evaluated.csv")['id'].isin(flaw_q)] for llm in llms}
```

```python
dfs['Qwen2.5-0.5B-Instruct'].head(3)
```

```python
cols = {
    "standard-answer": "standard-eval",
    "domain-answer": "domain-eval",
    "rdf-onto-answer": "rdf-onto-eval",
    "omn-onto-answer": "omn-onto-eval",
    "ofn-onto-answer": "ofn-onto-eval",
    "ttl-onto-answer": "ttl-onto-eval"
}

for llm, df in dfs.items():
    print(llm, end=" & ")
    for _, eval_col in cols.items():
        avg = df[eval_col].mean()*100
        print(str(avg)[:4], end=" & ")
    print("\\\\")
```
